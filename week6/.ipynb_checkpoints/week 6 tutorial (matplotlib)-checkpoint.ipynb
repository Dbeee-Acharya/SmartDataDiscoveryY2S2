```python

# Define number of points for the curve
num_points = 200

# Define an array of angles ranging from 0 to 2*pi
theta = np.linspace(0, 2*np.pi, num_points)

# Offset values to adjust the center and scale of the heart
x_offset = 0
y_offset = 0
scale = 1

# Define the heart curve equation (cardioid) with a little scaling factor
radius = 16 * (1 - np.cos( theta )**5) 

# Convert polar coordinates to cartesian coordinates
x = radius * np.cos( theta ) + x_offset
y = radius * np.sin( theta ) + y_offset

# Plot the heart shape
plt.plot(y, x, marker="*", color = "#fc03d7")

# Set aspect ratio for a better view (optional)
plt.gca().set_aspect('equal')

# Add labels and title (optional)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Heart")

# Show the plot
plt.show()
```

```python
x = np.random.randint(1,100,10)
y = np.random.randint(1,100,10)

# plt.figure(figsize=(2,3))
plt.title("Demo Matplotlib")
plt.xlabel('x axis', loc = 'right')  # location

plt.ylabel('y axis')

plt.plot(
    x, y, 
    marker = "*", 
    color = '#fc03d7', 
    markersize = 25, 
    markerfacecolor= '#a88aa4', 
    markeredgecolor="#401a3a")
```
