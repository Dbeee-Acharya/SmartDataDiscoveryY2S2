```python
def db_check():
    mydb = mysql.connector.connect(host='localhost', user='root',password='')
    
    if mydb:
        print("Connected")
        mycursor = mydb.cursor()
        mycursor.execute("show databases")
        
        for i in mycursor:
            print(i)
            
    else:
        print("Error")
        
db_check()
    
```

```python
def db_create():
    mydb = mysql.connector.connect(host='localhost', user='root',password='')
    
    if mydb:
        print("Connected")
        mycursor = mydb.cursor()
        mycursor.execute("CREATE DATABASE naya_database")
        
        for i in mycursor:
            print(i)
            
    else:
        print("Error")
        
db_create()
```

```python
def table_create():
    mydb = mysql.connector.connect(host='localhost', user='root',password='', database="naya_database")
    
    if mydb:
        print("Connected")
        mycursor = mydb.cursor()
        mycursor.execute("CREATE TABLE info(name VARCHAR(50), age INT)")
        
        print(mycursor)
            
    else:
        print("Error")
        
table_create()
```

```python
def table_insert():
    mydb = mysql.connector.connect(host='localhost', user='root',password='', database="naya_database")
    
    if mydb:
        print("Connected")
        mycursor = mydb.cursor()
        mycursor.execute("INSERT INTO info VALUES('guru', 22)")
        mydb.commit()
        
        print(mycursor)
            
    else:
        print("Error")

table_insert()
```

```python


def table_read():
    mydb = mysql.connector.connect(host='localhost', user='root',password='', database="naya_database")
    
    if mydb:
        print("Connected")
        mycursor = mydb.cursor()
        mycursor.execute("SELECT * FROM info")
        
        data = mycursor.fetchall()
        print(data)
        
        data = mycursor.fetchone()
        print(data)
        
        data = mycursor.fetchmany()
        print(data)
        
    else:
        print("Error")
        
table_read()
```
